```python
week = [ "Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for i in week:
  print(i)
```

```python
week = [ "Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for i in week:
  print(i)
  if i == "Thursday":
    break
```

```python
for x in range(6):
  print(x)
for x in range(2, 6):
  print(x)
```

```python
adjective = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adjective:
  for y in fruits:
    print(x, y)
```
